```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
road_data_dir = 'E:/data/road'
gps_data_dir = 'E:/data/gps'
```

```python
road_centroid = {
    1: (15.561914417289183, 16.85901078855188),
    2: (14.850897312327184, 16.27010669391355),
    4: (11.495837799354899, 13.314148102062967),
    7: (13.113253464736975, 13.73277648540468),
    9: (13.642726998935187, 13.649629122634497),
    32: (14.198731291192727, 15.275452778629077),
    41: (8.167021308334858, 9.946246687710023),
    304: (13.879089398416676, 14.388486307650279),
    35: (13.34104092128205, 13.640325795495965),
    331: (13.11827478284012, 13.59731304067158), 
}
```

```python
def get_road_data(road_data_dir,rd):
    df_road = pd.read_csv(road_data_dir + '/roaddb.csv')
    df_road = df_road[df_road['rd'] == rd].reset_index(drop=True)
    return df_road

def get_road_data_list(road_data_dir, rd_list):
    df_road = pd.read_csv(road_data_dir + '/roaddb.csv')
    df_road = df_road[df_road['rd'].isin(rd_list)].reset_index(drop=True)
    df_road['Road Segments'] = df_road.apply(lambda row: f"Highway No. {(row['rd'])}", axis=1)
    return df_road

def get_road_data_list_start_stop(road_data_dir, rd_list):
    df_road_tmp = []
    for rd in rd_list:
        if rd == 331:
            df_road = pd.read_csv(road_data_dir + '/roaddb.csv')
            df_road = df_road[(df_road['rd'] == rd) & (df_road['lat'] >= 13.09200) & (df_road['lat'] <= road_centroid[rd][1])].reset_index(drop=True)
            df_road['Road Segments'] = df_road.apply(lambda row: f"Highway No. {(row['rd'])}", axis=1)
        elif rd == 7:
            df_road = pd.read_csv(road_data_dir + '/roaddb.csv')
            df_road = df_road[(df_road['rd'] == rd) & (df_road['lat'] >= road_centroid[rd][0]) & (df_road['lat'] <= road_centroid[rd][1]) & (df_road['lon'] > 100.7886)].reset_index(drop=True)
            df_road = df_road[(df_road['lat'] < 13.56424) | (df_road['lat'] > 13.5985) | (df_road['lon'] < 100.9485) | (df_road['lon'] > 100.9539)].reset_index(drop=True)
            df_road = df_road[((df_road['lat'] < 13.2990) | (df_road['lat'] > 13.33408)) | ((df_road['lon'] < 100.989) | (df_road['lon'] > 100.9945))].reset_index(drop=True)
            df_road['Road Segments'] = df_road.apply(lambda row: f"Highway No. {(row['rd'])}", axis=1)
        elif rd == 304:
            df_road = pd.read_csv(road_data_dir + '/roaddb.csv')
            df_road = df_road[(df_road['rd'] == rd) & (df_road['lat'] >= road_centroid[rd][0]) & (df_road['lat'] <= road_centroid[rd][1]) & (df_road['lon'] > 100.5904)].reset_index(drop=True)
            df_road['Road Segments'] = df_road.apply(lambda row: f"Highway No. {(row['rd'])}", axis=1)
        elif rd == 9:
            df_road = pd.read_csv(road_data_dir + '/roaddb.csv')
            df_road = df_road[(df_road['rd'] == rd)].reset_index(drop=True)
            df_road = df_road[((df_road['lat'] > road_centroid[rd][0]) | (df_road['lat'] > road_centroid[rd][1])) | ((df_road['lon'] < 100.41451712861084) | (df_road['lon'] > 100.68650542608228))].reset_index(drop=True)
            df_road['Road Segments'] = df_road.apply(lambda row: f"Highway No. {(row['rd'])}", axis=1)
            # df_road
        else:
            df_road = pd.read_csv(road_data_dir + '/roaddb.csv')
            df_road = df_road[(df_road['rd'] == rd) & (df_road['lat'] >= road_centroid[rd][0]) & (df_road['lat'] <= road_centroid[rd][1])].reset_index(drop=True)
            df_road['Road Segments'] = df_road.apply(lambda row: f"Highway No. {(row['rd'])}", axis=1)
        df_road_tmp.append(df_road)
    result = pd.concat(df_road_tmp)
    return result
```
